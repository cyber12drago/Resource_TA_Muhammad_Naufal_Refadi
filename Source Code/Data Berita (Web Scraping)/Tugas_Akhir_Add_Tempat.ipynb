```python
provinces=[jawa_timur,jawa_tengah,jawa_barat,banten,di_yogyakarta,dki_jakarta,aceh,sumatera_utara,sumatera_barat,sumatera_selatan,riau]
provinces.extend((kep_riau,lampung,jambi,bengkulu,bangka_belitung,kalimantan_barat,kalimantan_timur,kalimantan_tengah,kalimantan_selatan,kalimantan_utara))
provinces.extend((sulawesi_utara,gorontalo,sulawesi_tengah,sulawesi_barat,sulawesi_selatan,sulawesi_tenggara,bali,nusa_tenggara_barat,nusa_tenggara_timur,maluku,maluku_utara,papua,papua_barat))    

for row in myresult:
    tags_word=''
    area=""
    kota=""
    cek=0
    for province in provinces:
        for tag_area in province:
            tag_berita = row[1].lower()
            cari = r"\b"+ tag_area +r"\b"
            hasil = re.search(cari, tag_berita)
            
            if str(hasil)!="None":
                if(tag_area== "maluku" and "maluku utara" in row[1].lower()):
                    break
                elif(tag_area== "papua" and "papua barat" in row[1].lower()):
                    break
                elif(tag_area== "riau" and "kepulauan riau" in row[1].lower()):
                    break
                else:
                    if(cek==1):
                        area+=", "
                    elif(kota==""):
                         for i in range(3,len(province)):
                            tag_berita1 = row[0].lower()
                            tag_berita2 = row[1].lower()     
                            cari = r"\b"+ province[i] +r"\b"
                            hasil1 = re.search(cari, tag_berita1)
                            hasil2 = re.search(cari, tag_berita2)
                            if(str(hasil1)!="None" or str(hasil2)!="None"):
                                kota=province[i]
                                break
                    
                    area+=province[0]
                    cek=1
                    break
                
    if(cek==0):
        for tag_area in nasional:
            tag_berita = row[1].lower()
            cari = r"\b"+ tag_area +r"\b"
            hasil = re.search(cari, tag_berita)
            if str(hasil)!="None":
                area+= "indonesia"
                cek=1
                break
                
    if(area!=""):
        if("'" not in str(row[0])):
            sql = "UPDATE news SET area = '"+area+ "' WHERE title = '"+str(row[0])+"'"    
            print(sql)
            mycursor.execute(sql)
            mydb.commit()
            if(kota!=""):
                sql = "UPDATE news SET kota = '"+kota+ "' WHERE title = '"+str(row[0])+"'"    
                print(sql)
                mycursor.execute(sql)
                mydb.commit()
            
        else:
            sql = 'UPDATE news SET area = "'+area+ '" WHERE title = "'+str(row[0])+'"'
            print(sql)
            mycursor.execute(sql)
            mydb.commit()
            if(kota!=""):
                sql = 'UPDATE news SET kota = "'+kota+ '" WHERE title = "'+str(row[0])+'"'  
                print(sql)
                mycursor.execute(sql)
                mydb.commit()
        
        

print(mycursor.rowcount, "record inserted.")
mycursor.close()
```
